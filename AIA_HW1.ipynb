```python
#由於missing value 不多，故直接清除，基本上不影響結果
datasettrain[['SepalWidth']] = datasettrain[['SepalWidth']].fillna(np.mean(datasettrain[['SepalWidth']]))
datasettrain[['SepalLength']] = datasettrain[['SepalLength']].fillna(np.mean(datasettrain[['SepalLength']]))
miss_value = datasettrain.isnull().sum().sort_values(ascending = False)
```

```python
from sklearn.model_selection import train_test_split
#建立model
X = datasettrain[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
y = datasettrain['type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=100)
#將training data 與 testing data 切成 7 : 3
```

```python
from sklearn import cross_validation, svm, preprocessing, metrics,neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC

clf = neighbors.KNeighborsClassifier()
clf_fit = clf.fit(X_train,y_train)
test_y_clfpredicted = clf.predict(X_test)
accuracyKNN = metrics.accuracy_score(y_test, test_y_clfpredicted) 
print("KNN acurracy:",accuracyKNN)
#使用KNN進行預測
```

```python
svc = svm.SVC()
svc_fit = svc.fit(X_train,y_train)
test_y_SVMpredicted = svc.predict(X_test)
accuracySVM = metrics.accuracy_score(y_test, test_y_SVMpredicted) 
print("SVN acurracy:",accuracySVM)
#使用SVM進行預測
```

```python
testdata = clf.predict(datasettest[['花萼長度','花萼寬度','花瓣長度','花瓣寬度']])
#由於預測出來的準確度都一樣，所以我就使用KNN將datatest.csv的數據輸入近來，並進行預測，來輸出是哪一種Iris
```
